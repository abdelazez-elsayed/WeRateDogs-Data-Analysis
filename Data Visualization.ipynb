```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
sns.set(rc={'figure.figsize':(200,20)})
% matplotlib inline
```

# Data Visualization and insights from WeRateDogs twitter account

```python
master_df = pd.read_csv('twitter_archive_master.csv')
```

```python
a=master_df.groupby(['the_dogtionary']).retweet_count.mean()[12::-1]
b=master_df.groupby(['the_dogtionary']).favorite_count.mean()[12::-1]
ind = np.arange(len(a))  # the x locations for the groups
width = 0.35
retweet_bars = plt.bar(ind, a, width, color='b', alpha=.7, label='Retweets')
like_bars = plt.bar(ind+width, b, width, color='r', alpha=.7, label='likes')
plt.ylabel('No. of retweets and likes')
plt.xlabel('breed')
plt.title('Average number of retweets and likes per breed')
locations = ind + width /2
#labels ['3', '4', '5', '6', '7', '8', '9']# xtick locations
sns.set(rc={'figure.figsize':(30,9)})

plt.xticks(locations,list(a.index))
plt.legend()
```

### According to charts doggo-puppo breed is the most liked one

```python
b=a['first_prediction'].value_counts().sort_values(ascending=False)[12::-1]
b.plot.bar(title='most retweeted breeds')
```
